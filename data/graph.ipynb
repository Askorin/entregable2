```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results.csv")
print(df.head())
```

```python
# Create a list to hold the expanded indices
expanded_idx = []

# Iterate over each row in the dataframe
for _, row in df.iterrows():
    # Append the 'idx' value 'count' times to the list
    expanded_idx.extend([row['idx']] * row['count'])
    
expanded_df = pd.DataFrame(expanded_idx, columns=['idx'])
```

```python
x = df["idx"]
y = df["count"]


# fig, ax = plt.subplots()
# ax.plot(x, y)
# plt.show()

# Using seaborn kdeplot to visualize the distribution
plt.figure(figsize=(10, 6))
sns.kdeplot(expanded_df['idx'], fill=True)
plt.xlabel('Expanded idx')
plt.title('KDE Plot of Expanded DataFrame')
plt.show()



```
