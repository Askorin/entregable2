```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Este código fue utilizado para plottear distribución de hashing ##

df = pd.read_csv("hashing_results.csv")
print(df.head())
```

```python
expanded_idx = []

for _, row in df.iterrows():
    # Append the 'idx' value 'count' times to the list
    expanded_idx.extend([row['idx']] * row['count'])
    
expanded_df = pd.DataFrame(expanded_idx, columns=['idx'])
```
